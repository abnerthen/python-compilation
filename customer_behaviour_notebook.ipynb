```python
def parse_json_column(cell):
    try:
        # Replacing single quotes with double quotes to make it valid JSON
        cell = cell.replace("'", '"')
        return json.loads(cell) if pd.notnull(cell) else None
    except json.JSONDecodeError:
        return None

def parse_cols(df, name):
    return df[name].apply(parse_json_column)
```

```python
df['device_new'] = df['device'].apply(lambda x: x.replace("'", '"'))
try: 
    df['device_new'] = df['device_new'].apply(lambda x: json.loads(x))
except json.JSONDecodeError:
    df['device_new'] = df['device_new'].apply(lambda x: ast.literal_eval(x))

df['device_new'].apply(type)
```
